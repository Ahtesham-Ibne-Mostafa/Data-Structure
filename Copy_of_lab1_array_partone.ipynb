```python
# Complete the functions defined in this cell

# Test 01: Shift Left k cell
def shift_left(source, k):
  # TO DO
  # Hint, You can write a function for left shift once and then use it
  for j in range(k):
    i=0
    for k in range(len(source)-1):
      source[i]=source[i+1]
      source[i+1]=0
      
      i+=1
  return source


# Test 02: Rotate Left k cell
def rotate_left(source, k):
  # TO DO
  # Hint, You can write a function for left rotate once and then use it
  for f in range(k):
    i=-1
    k=source[-1]
    for j in range(len(source)-1):
      source[i]=source[j]
      source[j]=0
      i+=1
    source[-2]=k
  return source


# Test 03: Shift Right k cell
def shift_right(source, k):
  # TO DO
  # Hint, You can write a function for right shift once and then use it
  for j in range(k):
    i=len(source)-1
    for k in range(len(source)-1):
      source[i]=source[i-1]
      source[i-1]=0
      
      i-=1
  return source



# Test 04: Rotate Right k cell
def rotate_right(source, k):
  # TO DO
  # Hint, You can write a function for right rotate once and then use it
  for f in range(k):
    i=0
    k=source[0]
    j=len(source)-1
    for d in range(len(source)-1):
      source[i]=source[j]
      source[j]=0
      i-=1
      j-=1
    source[1]=k
  return source


# Test 05: Remove an element from an array
def remove(source, idx):
  # TO DO
  i=idx
  for k in range(idx,len(source)-1):
    source[i]=source[i+1]
    source[i+1]=0
    
    i+=1
  return source



# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
  # TO DO
  j=len(source)-1
  while j>-1:
    if source[j]==element:
      i=j
      for k in range(j,len(source)-1):
        source[i]=source[i+1]
        source[i+1]=0
        
        i+=1
    j-=1
  print(source)
  return source



# Test 07: Splitting an Array
def split_array(a):
  # TO DO
  flag=False
  su1=0
  for i in range(len(a)):
    su1+=a[i]
    su2=0
    for j in range(i+1,len(a)):
      su2+=a[j]
    if su1==su2:
      flag=True
      break
  return flag


# Test 08: Max Bunch Count
def max_bunch(a):
  # TO DO
  f=0
  for i in range(len(a)):
    l=1
    for j in range(i+1,len(a)):
      if a[i]!=a[j]:
        break
      else:
        l+=1
    if l>f:
      f=l
  return f
```
